```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
%matplotlib inline
```

```python
df = pd.read_excel("Topics_stars10000_project_keyword.xlsx")
df
```

```python
topic_dic = {}
for i in df['topic_keyword']:
    i_list = i.replace("[","").replace("]","").replace("'","").split(",")
    for i in i_list:
        if len(i) == 0:
            continue
        if i not in topic_dic:
            topic_dic[i] = 1
        else:
            topic_dic[i] +=1
topic_dic = sorted(topic_dic.items(), key=lambda x: x[1], reverse=True)
topic_dic
```
