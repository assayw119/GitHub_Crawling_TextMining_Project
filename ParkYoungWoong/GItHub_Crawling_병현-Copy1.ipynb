```python
import requests
from bs4 import BeautifulSoup
import lxml
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url = "https://github.com/apple"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36"}
res= requests.get(url,headers=headers)
res.raise_for_status()
soup=BeautifulSoup(res.content,"html.parser")
```

```python
soup
```

```python
options = webdriver.ChromeOptions()
options.add_argument("headless")
browser = webdriver.Chrome("./chromedriver",options=options)
# browser = webdriver.Chrome("./chromedriver")
browser.get(url)
soup = BeautifulSoup(browser.page_source,'html')
```

```python
repo_num = soup.find('span', attrs={'class':'Counter js-profile-repository-count'}).get_text()
repo_num
```

```python
# 브랜치 수(파생 프로젝트 수)
branch_num = soup.find("div",attrs={"class":"flex-self-center ml-3 flex-self-stretch d-none d-lg-flex flex-items-center lh-condensed-ultra"}).strong.get_text()
branch_num
```

```python
# 풀 리퀘스트 수(코드 병합 요구 수)
pull_requests_num = soup.find("a",attrs={"id":"pull-requests-tab"}).find("span",attrs={"class":"Counter"}).get_text()
pull_requests_num
```

```python
# 커밋 수(수정 요구수)
commit_num = soup.find("li",attrs={"class":"ml-0 ml-md-3"}).strong.get_text()
commit_num
```

```python
# 스타 수(북마크 수, 인기도)
star_num = soup.find("ul",attrs={"class":"pagehead-actions flex-shrink-0 d-none d-md-inline"}).find("a",attrs={"class":"social-count js-social-count"}).get_text()
star_num = star_num.replace('\t', '').replace('\n', '').strip()
star_num
```

```python
# 토픽(개발언어의 주제)
topic = soup.find("div",attrs={"class":"BorderGrid-cell"}).find_all("a",attrs={"class":"topic-tag topic-tag-link"})
topics=[i.get_text().replace("\n","").replace("\t","").strip() for i in topic]
topics
```

```python
# 릴리즈 정보(버전 정보)
release_info = soup.find("div",attrs={"class":"BorderGrid BorderGrid--spacious"}).find("span",attrs={"class":"css-truncate css-truncate-target text-bold mr-2"}).get_text().strip()
release_info
```

```python
# 작성 프로그래밍 언어 종류
program_lang = []
# soup.find("div",attrs={"class":"BorderGrid BorderGrid--spacious"}).find("ul",attrs={"class":"list-style-none"}).get_text()
for i in soup.find_all("a",attrs={"class":"d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"}):
    program_lang.append(i.find("span",attrs={"class":"color-text-primary text-bold mr-1"}).get_text())
if soup.find("span",attrs={"class":"d-inline-flex flex-items-center flex-nowrap text-small mr-3"}).span:
    program_lang.append(soup.find("span",attrs={"class":"d-inline-flex flex-items-center flex-nowrap text-small mr-3"}).span.get_text())
program_lang
```

```python
# read me 정보(코드 개요 텍스트)
readme = soup.find("div",attrs={"data-target":"readme-toc.content"}).article.get_text()
readme.replace("\n","").replace("  ","")
```
