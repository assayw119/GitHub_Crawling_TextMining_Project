```python
import re
import requests
from bs4 import BeautifulSoup
import lxml
import time
```

```python

result = []
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36"}
for n in range(1,25):
    print('< {} page >'.format(n))
#     params = { 
#         'p' : n
#     }
    url = 'https://github.com/search?p={}&q=stars%3A%3E10000&type=Repositories'.format(n)
    

    response= requests.get(url,headers=headers)
    response.raise_for_status()
    soup=BeautifulSoup(response.text,"lxml")
    
    contents = soup.find('ul',attrs={'class':'repo-list'})
    topic = contents.find_all('a', attrs={'class':'v-align-middle'})
    if response.status_code == 429:
        print(int(response.headers["Retry-After"]))
        time.sleep(int(response.headers["Retry-After"]))
    for i in topic:
        print(i.text)
    print()
```

```python
url_list = []
for n in range(1,101):
    url = 'https://github.com/search?p={}&q=stars%3A%3E10000&type=Repositories'.format(n)
    url_list.append(url)

url_list
    
```

```python
for i in range(len(url_list)):
    url_ = url_list[i]
    resp = requests.get(url_)
    soup = BeautifulSoup(resp.content, 'html.parser')
    contents = soup.find('div', attrs={'class':'px-2'}).find('ul')
    
    for j in contents.find_all('a'):
        print(j.text)
```
